```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from PIL import Image
```

```python
#content
text= open('F:\\Bhaiya\\Batman_wordcloud-master\\batman.txt','r').read()
```

```python
#appearence
custum_mask = np.array(Image.open('F:\\Bhaiya\\Batman_wordcloud-master\\cloud.png'))
wc=WordCloud(background_color="white", stopwords=STOPWORDS, mask=custum_mask,contour_width=3, contour_color='black')
wc.generate(text)
```

```python
#plotting
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
